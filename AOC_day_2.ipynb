```python
import numpy as np
```

```python
rounds = []
with open('input') as input:
  for line in input:
    round = [line[0], line[2]]
    rounds.append(round)
```

```python
def round_score(round):
  score = 0
  if (round[1] == 'X'):
    score += 1
    if round[0] == 'A':
      score += 3
    elif round[0] == 'C':
      score += 6
  elif (round[1] == 'Y'):
    score += 2
    if round[0] == 'B':
      score += 3
    elif round[0] == 'A':
      score += 6
  elif (round[1] == 'Z'):
    score += 3
    if round[0] == 'C':
      score += 3
    elif round[0] == 'B':
      score += 6
  return score
  
```

```python
def total_score(input):
  total = 0
  for round in input:
    total += round_score(round)
  return total
```

```python
final_score = total_score(rounds)
print(final_score)
```

```python
def round_score2(round):
  # X lose # A Rock
  # Y draw # B Paper
  # Z win  # C scissors
  score = 0

  if (round[1] == 'X'):
    if (round[0] == 'A'):
      score += 3
    elif (round[0] == 'B'):
      score += 1
    else:
      score += 2
  elif (round[1] == 'Y'):
    score += 3
    if (round[0] == 'A'):
      score += 1
    elif (round[0] == 'B'):
      score += 2
    else:
      score += 3
  elif (round[1] == 'Z'):
    score += 6
    if (round[0] == 'A'):
      score += 2
    elif (round[0] == 'B'):
      score += 3
    else:
      score += 1
  return score
```

```python
def total_score2(input):
  total = 0
  for round in input:
    total += round_score2(round)
  return total
```

```python
final_score2 = total_score2(rounds)
print(final_score2)
```
