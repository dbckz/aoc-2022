```python
directions = []
with open('input') as f:
  for line in f:
    directions.append(line.strip().split(' '))

directions = [[x[0], int(x[1])] for x in directions]
print(directions)
```

```python
dirs_string = ''
for d in directions:
  dirs_string += d[0] * d[1]
print(dirs_string)
```

```python
def run_direction(dirs_string):
  lead_coords = (0,0)
  follow_coords = (0,0)
  visited = set([(0,0)])

  for c in dirs_string:
    if c == 'U':
      lead_coords = (lead_coords[0], lead_coords[1] + 1)
      if (lead_coords[1] == follow_coords[1] + 2):
        follow_coords = (lead_coords[0], follow_coords[1] + 1)
        visited.add(follow_coords)
        continue
    if c == 'D':
      lead_coords = (lead_coords[0], lead_coords[1] - 1)
      if (lead_coords[1] == follow_coords[1] - 2):
        follow_coords = (lead_coords[0], follow_coords[1] - 1)
        visited.add(follow_coords)
        continue
    if c == 'L':
      lead_coords = (lead_coords[0] - 1, lead_coords[1])
      if (lead_coords[0] == follow_coords[0] - 2):
        follow_coords = (follow_coords[0] - 1, lead_coords[1])
        visited.add(follow_coords)
        continue
    if c == 'R':
      lead_coords = (lead_coords[0] + 1, lead_coords[1])
      if (lead_coords[0] == follow_coords[0] + 2):
        follow_coords = (follow_coords[0] + 1, lead_coords[1])
        visited.add(follow_coords)
        continue
  return(visited)

print(f"Answer 1: {len(run_direction(dirs_string))}")

```

```python
def find_direction(o, n):
  o_x = o[0]
  o_y = o[1]
  n_x = n[0]
  n_y = n[1]

  # up left
  if (n_x == o_x - 1) and (n_y == o_y + 1):
    return 'Q'
  # up
  if (n_x == o_x) and (n_y == o_y + 1):
    return 'U'
  # up right
  if (n_x == o_x + 1) and (n_y == o_y + 1):
    return 'P'
  # right
  if (n_x == o_x + 1) and (n_y == o_y):
    return 'R'
  # down right
  if (n_x == o_x + 1) and (n_y == o_y - 1):
    return 'M'
  # down
  if (n_x == o_x) and (n_y == o_y - 1):
    return 'D'
  # down left
  if (n_x == o_x - 1) and (n_y == o_y - 1):
    return 'Z'
  # left
  if (n_x == o_x - 1) and (n_y == o_y):
    return 'L'
  raise Exception(f"Coords: {n}, {o}")

def run_direction2(dirs_string):
  lead_coords = (0,0)
  follow_coords = (0,0)
  visited = set([(0,0)])
  follow_string = ''

  for c in dirs_string:
    print(c)
    if c == 'U':
      lead_coords = (lead_coords[0], lead_coords[1] + 1)
      if (lead_coords[1] == follow_coords[1] + 2):
        new_follow_coords = (lead_coords[0], follow_coords[1] + 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'D':
      lead_coords = (lead_coords[0], lead_coords[1] - 1)
      if (lead_coords[1] == follow_coords[1] - 2):
        new_follow_coords = (lead_coords[0], follow_coords[1] - 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'L':
      lead_coords = (lead_coords[0] - 1, lead_coords[1])
      if (lead_coords[0] == follow_coords[0] - 2):
        new_follow_coords = (follow_coords[0] - 1, lead_coords[1])
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'R':
      lead_coords = (lead_coords[0] + 1, lead_coords[1])
      print(lead_coords)
      if (lead_coords[0] == follow_coords[0] + 2):
        new_follow_coords = (follow_coords[0] + 1, lead_coords[1])
        print(follow_coords)
        print(new_follow_coords)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'P':
      lead_coords = (lead_coords[0] + 1, lead_coords[1] + 1)
      if (lead_coords[0] == follow_coords[0] + 2) and (lead_coords[1] == follow_coords[1] + 2):
        new_follow_coords = (follow_coords[0] + 1, follow_coords[1] + 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      elif (lead_coords[0] == follow_coords[0] + 1) and (lead_coords[1] == follow_coords[1] + 2):
        new_follow_coords = (follow_coords[0] + 1, follow_coords[1] + 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      elif (lead_coords[0] == follow_coords[0]) and (lead_coords[1] == follow_coords[1] + 2):
        new_follow_coords = (follow_coords[0], follow_coords[1] + 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      elif (lead_coords[0] == follow_coords[0] + 2) and (lead_coords[1] == follow_coords[1]):
        new_follow_coords = (follow_coords[0] + 1, follow_coords[1])
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      elif (lead_coords[0] == follow_coords[0] + 2) and (lead_coords[1] == follow_coords[1] + 1):
        new_follow_coords = (follow_coords[0] + 1, follow_coords[1] + 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'Q':
      lead_coords = (lead_coords[0] - 1, lead_coords[1] + 1)
      if (lead_coords[0] == follow_coords[0] - 2):
        new_follow_coords = (follow_coords[0] - 1, follow_coords[1] + 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'Z':
      lead_coords = (lead_coords[0] - 1, lead_coords[1] - 1)
      if (lead_coords[0] == follow_coords[0] - 2):
        new_follow_coords = (follow_coords[0] - 1, follow_coords[1] - 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
    if c == 'M':
      lead_coords = (lead_coords[0] + 1, lead_coords[1] - 1)
      if (lead_coords[0] == follow_coords[0] + 2):
        new_follow_coords = (follow_coords[0] + 1, follow_coords[1] - 1)
        follow_string += find_direction(follow_coords, new_follow_coords)
        follow_coords = new_follow_coords
        visited.add(follow_coords)
      continue
  return follow_string, visited
```
